## Statistical Summary

```python
df["Price"] = df["Price"].str.replace("£", "", regex=False)

df["Price"] = pd.to_numeric(df["Price"])

print(df.dtypes)
```

```python
df.describe()
```

## Questions for Analysis

1. What is the average price of books?
2. Which rating category appears most frequently?
3. What are the highest and lowest priced books?
4. Are highly rated books more expensive?
5. Are there any unusual price values in the dataset?

## Average Book Price

```python
print("Average Price:", df["Price"].mean())
```

## Highest and Lowest Book Prices

```python
print("Highest Price:")
print(df[df["Price"] == df["Price"].max()])

print("\nLowest Price:")
print(df[df["Price"] == df["Price"].min()])
```

## Rating Distribution

```python
print(df["Rating"].value_counts())
```

## Price Distribution Visualization

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Price"], bins=10)

plt.title("Distribution of Book Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")

plt.show()
```

## Book Ratings Visualization

```python
sns.countplot(x="Rating", data=df)

plt.title("Book Ratings Count")

plt.show()
```

## Detecting Potential Anomalies

```python
print(df[df["Price"] > 55])
```

## Insights

- Most books fall within a moderate price range.
- Five-star and one-star books appear frequently.
- Some books have significantly higher prices than average.
- The dataset does not contain missing values.
